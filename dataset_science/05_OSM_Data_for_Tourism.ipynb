# 05 – OSM Data for Tourism

This notebook gives download/access points for full Angola OpenStreetMap data from Geofabrik and HOTOSM.

```python
from pathlib import Path
import pandas as pd
import requests, json, os, re
ROOT = Path.cwd()
if ROOT.name == 'starter_notebooks':
    ROOT = ROOT.parent
DATA = ROOT / 'data'
RAW = DATA / 'raw'
SAMPLE = DATA / 'sample'
RAW.mkdir(parents=True, exist_ok=True)
print('Project root:', ROOT)
```

```python
# Direct access links
links = {
    'Geofabrik Angola page': 'https://download.geofabrik.de/africa/angola.html',
    'Geofabrik latest OSM PBF': 'https://download.geofabrik.de/africa/angola-latest.osm.pbf',
    'Geofabrik latest free shapefile zip': 'https://download.geofabrik.de/africa/angola-latest-free.shp.zip',
    'HOTOSM Angola POIs': 'https://data.humdata.org/dataset/hotosm_ago_points_of_interest',
    'HOTOSM Angola Roads': 'https://data.humdata.org/dataset/hotosm_ago_roads',
}
pd.DataFrame([{'name':k,'url':v} for k,v in links.items()])
```
