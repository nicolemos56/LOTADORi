# 06 – SDG and Tourism Impact Dashboard

Uses the World Bank API for Angola indicators relevant to sustainable tourism and SDGs.

```python
from pathlib import Path
import pandas as pd
import requests, json, os, re
ROOT = Path.cwd()
if ROOT.name == 'starter_notebooks':
    ROOT = ROOT.parent
DATA = ROOT / 'data'
RAW = DATA / 'raw'
SAMPLE = DATA / 'sample'
RAW.mkdir(parents=True, exist_ok=True)
print('Project root:', ROOT)
```

```python
indicators = {
    'SP.POP.TOTL':'Population total',
    'SL.UEM.TOTL.ZS':'Unemployment total (%)',
    'ST.INT.ARVL':'International tourism arrivals',
    'AG.LND.FRST.ZS':'Forest area (% of land area)',
    'EG.ELC.ACCS.ZS':'Access to electricity (% of population)',
    'IT.NET.USER.ZS':'Individuals using the Internet (% of population)'
}
records=[]
for code, label in indicators.items():
    url=f'https://api.worldbank.org/v2/country/AGO/indicator/{code}?format=json&per_page=20000'
    data=requests.get(url, timeout=60).json()
    for row in data[1] if len(data)>1 and data[1] else []:
        records.append({'indicator_code':code,'indicator':label,'year':int(row['date']),'value':row['value']})
wb = pd.DataFrame(records).dropna(subset=['value'])
wb.head()
```

```python
latest = wb.sort_values('year').groupby('indicator').tail(1).sort_values('indicator')
latest
```

```python
import matplotlib.pyplot as plt
sel = wb[wb['indicator_code'].eq('ST.INT.ARVL')].sort_values('year')
plt.figure(figsize=(10,4))
plt.plot(sel['year'], sel['value'], marker='o')
plt.title('Angola - International Tourism Arrivals')
plt.xlabel('Year')
plt.ylabel('Arrivals')
plt.grid(True)
plt.show()
```
