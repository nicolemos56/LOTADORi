# 02 – Interactive Tourism Map with Folium

Uses the ready sample dataset first. Teams can later replace it with HOTOSM POIs or boundary data.

```python
from pathlib import Path
import pandas as pd
import requests, json, os, re
ROOT = Path.cwd()
if ROOT.name == 'starter_notebooks':
    ROOT = ROOT.parent
DATA = ROOT / 'data'
RAW = DATA / 'raw'
SAMPLE = DATA / 'sample'
RAW.mkdir(parents=True, exist_ok=True)
print('Project root:', ROOT)
```

```python
import folium

df = pd.read_csv(SAMPLE / 'angola_tourism_attractions_sample.csv')
center = [df['latitude'].mean(), df['longitude'].mean()]
m = folium.Map(location=center, zoom_start=6)
for _, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"<b>{row['name']}</b><br>{row['category']}<br>{row['province']}",
        tooltip=row['name']
    ).add_to(m)
m
```

```python
# Save map as HTML
out = ROOT / 'data' / 'processed' / 'sample_tourism_map.html'
m.save(out)
out
```
