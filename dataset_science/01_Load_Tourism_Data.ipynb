# 01 – Load Tourism Data

**Dataset:** HOTOSM Angola Points of Interest + sample Angola tourism attractions.

This notebook lets participants start immediately with the sample file, then optionally download the larger OpenStreetMap/HDX dataset.

```python
from pathlib import Path
import pandas as pd
import requests, json, os, re
ROOT = Path.cwd()
if ROOT.name == 'starter_notebooks':
    ROOT = ROOT.parent
DATA = ROOT / 'data'
RAW = DATA / 'raw'
SAMPLE = DATA / 'sample'
RAW.mkdir(parents=True, exist_ok=True)
print('Project root:', ROOT)
```

```python
# Load the small ready-to-use sample dataset
sample_path = SAMPLE / 'angola_tourism_attractions_sample.csv'
df = pd.read_csv(sample_path)
df.head()
```

```python
# Quick analysis
df.groupby(['province','category']).size().reset_index(name='count').sort_values('count', ascending=False)
```

```python
# Optional: discover HDX resources for Angola POIs
api_url = 'https://data.humdata.org/api/3/action/package_show?id=hotosm_ago_points_of_interest'
meta = requests.get(api_url, timeout=60).json()
resources = pd.DataFrame(meta['result']['resources'])[['name','format','url']]
resources.head(20)
```

```python
# Optional: download a CSV/GeoJSON resource if available
out_dir = RAW / 'hotosm_ago_points_of_interest'
out_dir.mkdir(exist_ok=True)
preferred = resources[resources['format'].fillna('').str.lower().str.contains('csv|geojson|json', regex=True)]
preferred[['name','format','url']].head()
```
