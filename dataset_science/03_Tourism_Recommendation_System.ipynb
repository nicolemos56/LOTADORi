# 03 – Simple Tourism Recommendation System

A beginner-friendly recommendation model using category, province and text description.

```python
from pathlib import Path
import pandas as pd
import requests, json, os, re
ROOT = Path.cwd()
if ROOT.name == 'starter_notebooks':
    ROOT = ROOT.parent
DATA = ROOT / 'data'
RAW = DATA / 'raw'
SAMPLE = DATA / 'sample'
RAW.mkdir(parents=True, exist_ok=True)
print('Project root:', ROOT)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_csv(SAMPLE / 'angola_tourism_attractions_sample.csv')
df['text'] = df[['name','category','province','municipality','description']].astype(str).agg(' '.join, axis=1)
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['text'])

def recommend(query, top_n=5):
    q = vectorizer.transform([query])
    scores = cosine_similarity(q, X).ravel()
    out = df.copy()
    out['score'] = scores
    return out.sort_values('score', ascending=False)[['name','category','province','description','score']].head(top_n)

recommend('beach coastal nature surfing near Luanda')
```

```python
recommend('culture museum heritage history')
```
