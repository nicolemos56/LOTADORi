```python
#Passo 1: Carregar e Inspecionar os Dados Brutos
import pandas as pd
import geopandas as gpd

#Leitura de dataset bruto
df = pd.read_csv('../data_bruto/dados_brutos_turismo.csv')
print(df.head())
print(df.info())
```

```python
# 1. Verificar se as colunas essenciais existem e limpar valores nulos no nome
df_limpo = df.dropna(subset=['name'])

# 2. Mapear as tuas colunas atuais para o padrão que a base de dados espera
df_limpo = df_limpo.rename(columns={
    'name': 'name',
    'type': 'category',       # O 'type' do teu dataset serve perfeitamente como categoria
    'description': 'description'
})

# 3. Como este dataset parece não ter latitude/longitude explícitas (ou estão dentro de outra coluna), 
# se precisares de preencher província ou coordenadas para teste, podes definir valores padrão por enquanto:
df_limpo['province'] = 'Malanje'  # Ou a província correspondente
df_limpo['latitude'] = -9.5367    # Exemplo de coordenada padrão (ex: Malanje)
df_limpo['longitude'] = 16.3444

# 4. Selecionar apenas as colunas que o modelo Place do backend reconhece
colunas_finais = ['name', 'category', 'province', 'latitude', 'longitude', 'description']
df_limpo = df_limpo[[c for c in colunas_finais if c in df_limpo.columns]]

# 5. Exportar o resultado pronto para seeding
df_limpo.to_csv('locais_turisticos_tratados.csv', index=False)
print("Dados tratados com sucesso! Ficheiro 'locais_turisticos_tratados.csv' gerado.")
```
